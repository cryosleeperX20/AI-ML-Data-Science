```python
df_num = df_houseprice.select_dtypes(include = [np.number])
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_target = df_houseprice['Price']
df_num = df_num.drop('Price', axis = 1)
df_num.head()
```

```python
# initiatate the standard scaler
ss = StandardScaler()
```

```python
# transform the data 
Scale = ss.fit_transform(df_num)
Scale
```

```python
df_Scale = pd.DataFrame(Scale, columns = df_num.columns)
df_Scale = df_Scale.reset_index()
df_Scale
```

```python
#make a dataframe indipendent variable 
df_independant = pd.concat([df_Scale,dummy_var],axis = 1)
df_independant
```

```python
#split the data for traingand testing a data
import statsmodels
import statsmodels.api as sm
```

```python
df_independant = sm.add_constant(df_independant)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_independant, 
                                                    df_target, 
                                                    random_state = 10, 
                                                    test_size = 0.2)
```

```python
print('X_train',X_train.shape)
print('y_train',y_train.shape)
print('X_test',X_test.shape)
print('y_test',y_test.shape)
```

```python
X_train.columns
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# create a Linear Regration model
model = LinearRegression()

# fit model to train data
model.fit(X_train, y_train)

# precticle on the testing data
predictions = model.predict(X_test)

mse = mean_squared_error(y_test,predictions)
print('Mean Squred Error is :',mse)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
```

```python
mae = mean_absolute_error(y_test,predictions)
mae
```

```python
rmse = root_mean_squared_error(y_test,predictions)
rmse
```
